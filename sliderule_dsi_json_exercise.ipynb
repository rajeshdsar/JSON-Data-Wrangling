## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
#Read the JSON file as a DataFrame
df_wbp = pd.read_json('data/world_bank_projects.json')
```

```python
#see whats in the file
print(df_wbp.shape)
print(df_wbp.columns)
print(df_wbp.head(2))
```

```python
#Find the 10 countries with most projects
df_wbp.countrycode.value_counts().head(10)
```

```python
#Find the top 10 major project themes (using column 'mjtheme_namecode')
#step 1 - normalize and load the JSON into a dataframe, load the country name and the fields in the major theme
major_themes = json_normalize(json.load((open('data/world_bank_projects.json'))),
                              'mjtheme_namecode',['countrycode','countryshortname'])
#step 2 - find the 10 major project themes by checking the counts of 'code' column
major_themes.code.value_counts().head(10)
```

```python
#In 2. above you will notice that some entries have only the code and the name is missing. 
#Create a dataframe with the missing names filled in.

#step 1 - sort the dataframe by 'code' column
major_themes = major_themes.sort_values('code')

#step 2 - some of the columns have '', replace that with NaN, and use forward fill to update the NaN values
major_themes.name = major_themes.name.replace('',method='ffill')

#step 3 - display the dataframe to check if it looks right
major_themes
```

```python
#step 4 - use group by to list the top 10 major themes
major_themes.groupby(['code','name']).size().sort_values(ascending=False).head(10)
```
